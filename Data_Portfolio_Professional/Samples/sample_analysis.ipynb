```python
# Data Portfolio Analysis Examples
# Save this file, then rename to .ipynb

print("ðŸ“Š PORTFOLIO DATA ANALYSIS")
print("=" * 50)

# Example 1: E-commerce Analysis
print("\n1. E-COMMERCE MARGIN ANALYSIS")
print("-" * 30)

import pandas as pd
import matplotlib.pyplot as plt

# Load e-commerce data
ecom_df = pd.read_csv('../datasets/ecommerce/ecommerce_data.csv')
print(f"âœ“ Loaded {len(ecom_df)} products")
print(f"â€¢ Average margin: {ecom_df['Margin'].mean():.1f}%")
print(f"â€¢ Highest price: ${ecom_df['Price'].max():.2f}")
print(f"â€¢ Categories: {ecom_df['Category'].nunique()} types")

# Top 5 products by margin
print("\nTop 5 Products by Margin:")
top_products = ecom_df.nlargest(5, 'Margin')[['Product_Name', 'Category', 'Margin']]
print(top_products.to_string(index=False))

# Example 2: Real Estate Analysis
print("\n\n2. REAL ESTATE INVESTMENT ANALYSIS")
print("-" * 30)

re_df = pd.read_csv('../datasets/real_estate/realestate_data.csv')
print(f"âœ“ Loaded {len(re_df)} properties")
print(f"â€¢ Average price: ${re_df['List_Price'].mean():,.0f}")
print(f"â€¢ Average cap rate: {re_df['Cap_Rate'].mean():.1f}%")

# Best investment cities
print("\nBest Cities for Investment (by Cap Rate):")
city_roi = re_df.groupby('City')['Cap_Rate'].mean().sort_values(ascending=False)
for city, rate in city_roi.head(3).items():
    print(f"  {city}: {rate:.1f}% ROI")

# Example 3: Quick Healthcare Stats
print("\n\n3. HEALTHCARE COST SUMMARY")
print("-" * 30)

health_df = pd.read_csv('../datasets/healthcare/healthcare_data.csv')
print(f"âœ“ Loaded {len(health_df)} patient records")
print(f"â€¢ Average treatment cost: ${health_df['Treatment_Cost'].mean():,.0f}")
print(f"â€¢ Average stay: {health_df['Length_of_Stay'].mean():.1f} days")

print("\n" + "=" * 50)
print("âœ… ANALYSIS COMPLETE")
print("These datasets are ready for business intelligence.")
print("Use for: Pricing strategies, Investment decisions, Cost analysis")
```
